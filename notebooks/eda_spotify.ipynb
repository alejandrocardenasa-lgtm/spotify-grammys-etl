```python
plt.figure()
plt.boxplot(spotify["popularity"].dropna())
plt.title("Boxplot of Popularity")
plt.ylabel("Popularity")
plt.show()
```

```python
plt.figure()
plt.boxplot(spotify["duration_ms"].dropna())
plt.title("Boxplot of Duration")
plt.ylabel("Duration (ms)")
plt.show()
```

```python
plt.figure()
plt.hist(spotify["popularity"].dropna(), bins=30)
plt.title("Distribution of Popularity")
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.hist(spotify["duration_ms"].dropna(), bins=30)
plt.title("Distribution of Song Duration")
plt.xlabel("Duration (ms)")
plt.ylabel("Frequency")
plt.show()
```

```python
top_artists = spotify["artists"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(top_artists.index, top_artists.values)
plt.title("Top 10 Artists")
plt.xticks(rotation=90)
plt.ylabel("Count")
plt.show()
```

```python
top_genres = spotify["track_genre"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(top_genres.index, top_genres.values)
plt.title("Top 10 Genres")
plt.xticks(rotation=90)
plt.ylabel("Count")
plt.show()
```
